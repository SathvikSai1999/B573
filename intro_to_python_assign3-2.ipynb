```python
file_path = "/Users/sathviksai/Downloads/"
file_name = file_path + "chr1_GL383518v1_alt.fa"

with open(file_name, 'r') as file:  
    next(file)  #helps skip the first line of the file (header)
    sequence = file.read().replace('\n', '') #Removes newlines within the fasta sequence

print(sequence) 
```

```python
#1. 
# Print the 10th letter 
letter_10th = sequence[9]
print("10th Letter is:", letter_10th)

# Print the 758th letter 
letter_758 = sequence[757]
print("758th Letter is:", letter_758)
```

```python
#converting lower case to upper case
all_caps_dict = {'a': 'A', 't': 'T', 'g': 'G', 'c': 'C'}
all_caps_sequence = "".join(all_caps_dict.get(char, char) for char in sequence)
print("all_caps_sequence:",all_caps_sequence)
```

```python
#complement strand
complement_dict = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
complementary_strand = "".join(complement_dict.get(char, char) for char in all_caps_sequence) 
print("Complementary strand is:", complementary_strand)
```

```python
#reverse complement strand
reverse_complement_strand = complementary_strand[::-1]
print("reverse complement strand is:", reverse_complement_strand)
```

```python
#2. 
# Print the 79th letter 
print("79th letter:", reverse_complement_strand[78]) 

# Print the 500th through 800th letters 
print("500th-800th letters:", reverse_complement_strand[499:800]) 
```

```python
#Number of nucleotides:
num_nucleotides = len(reverse_complement_strand)
print("Number of nucleotides:", num_nucleotides)
```

```python
#3
def count_nucleotides_in_kilobases(sequence, kilobase=1000):
    results = {}
    for i in range(0, len(sequence), kilobase):
        chunk = sequence[i:i + kilobase]
        counts = {'A': 0, 'T': 0, 'G': 0, 'C': 0}
        for letter in chunk:
            counts[letter] += 1

        # Store the kilobase number (starting from 1)
        kilobase_number = i // kilobase + 1  
        results[kilobase_number] = counts 
    return results


results = count_nucleotides_in_kilobases(all_caps_sequence)

for kilobase_number, counts in results.items():
    print(f"kilobase {kilobase_number}: {counts}") 

```
